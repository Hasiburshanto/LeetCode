```python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        length = len(nums)
        result = []

        for i in range (length-3):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            for j in range(i+1,length-2):
                if j > i+1 and nums[j] == nums[j-1]:
                    continue
                left = j + 1
                right = length -1
                while (left < right):
                    total = nums[i] + nums[j] + nums[left] + nums[right]
                    print(nums[i],nums[j], nums[left], nums[right])
                    print(total)
                    if ( total == target ) :
                        result.append([nums[i] , nums[j] , nums[left] , nums[right]])
                        
                        while left < right and nums[left] == nums[left+1] :
                            left = left + 1
                        while left < right and nums[right] == nums[right-1] :
                            right = right - 1
                        
                        left += 1
                        right -= 1
                        
                    elif total < target:
                        left += 1
                    else:
                        right -= 1

        return result         

# Testing the Solution
solution = Solution()

# Example 1
nums1 = [5,5,3,5,1,-5,1,-2]
target1 = 4
output1 = solution.fourSum(nums1, target1)
print(f"Input: nums = {nums1}, target = {target1}\nOutput: {output1}\n")
#-5,-2,1,1,3,5,5,5
#[[-5,1,3,5]]
# Example 2
#nums2 = [2,2,2,2,2]
#target2 = 8
#output2 = solution.fourSum(nums2, target2)
#print(f"Input: nums = {nums2}, target = {target2}\nOutput: {output2}\n")
```
